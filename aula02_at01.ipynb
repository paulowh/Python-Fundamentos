```python
#operadores....

num1 = input('digite o primeiro numero: ')
num2 = input('digite o segundo numero: ')

soma = int(num1) + int(num2)
print('o valor da soma {}'.format(soma))

subtracao = int(num1) - int(num2)
print('o valor da subtração {}'.format(subtracao))

multiplicacao = int(num1) * int(num2)
print('o valor da multiplicação {}'.format(multiplicacao))

divisao = int(num1) / int(num2)
print(type(divisao))
print('o valor da divisão {:.3f}'.format(divisao))

# altura = 1.70
# altura == 1.60 -> false
# altura > 1.60 -> true
# altura < 1.60 -> false
# altura >= 1.60 -> true
# altura <= 1.60 -> false
# altura != 1.60 -> true
```

```python
nota1 = input('Qual sua primeira nota: ')
nota2 = input('Qual sua segunda nota: ')
nota3 = input('Qual sua terceira nota: ')
nota4 = input('Qual sua quarta nota: ')

soma = int(nota1) + int(nota2) + int(nota3)+ int(nota4)
print(type(soma))

media = soma / 4
print(type(media))

#if pode esta sozinho
#elif não pode existir sem o if
#else é opcional

if float(media) > 8:
  print('você passou de ano, parabens')
elif float(media) >= 5:
  print('você pode fazer recuperação')
else:
  print('você foi convidado a participar mais um ano')
```
