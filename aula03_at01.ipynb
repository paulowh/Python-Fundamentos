```python
#receber um numero do usuario e verificar se o numero é primo

numero = int(input('digite um numero: '))
contador = 1
ehPrimo = 0

while contador < numero + 1:
  print(contador)
  if numero % contador == 0:
    ehPrimo += 1

  contador += 1

if ehPrimo == 2:
  print('o numero {} é PRIMO'.format(numero))
else:
  print('o numero {} NÃO PRIMO'.format(numero))

```
